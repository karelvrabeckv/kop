```python
import constants as c
import tester as t

bf_results = t.run(c.BF, c.ALL_INSTANCES, 20, c.NR_INST_FILE_PATH_PREFIX)
bf_better_results = t.run(c.BF_BETTER, c.ALL_INSTANCES, 20, c.NR_INST_FILE_PATH_PREFIX)
bb_results = t.run(c.BB, c.ALL_INSTANCES, 35, c.NR_INST_FILE_PATH_PREFIX)
```

```python
import helpers as h

bf_average_times = []
for bf_result in bf_results:
  time = h.getAverageTime(bf_result[1])
  bf_average_times.append(time)
print(bf_average_times)

bf_max_times = []
for bf_result in bf_results:
  time = h.getMaxTime(bf_result[0], "configurations")
  bf_max_times.append(time)
print(bf_max_times)
```

```python
import numpy as np
import matplotlib.pyplot as plt

barWidth = 1.0
fig = plt.subplots(figsize = (12, 6))
br1 = np.arange(15)
br2 = [x + barWidth for x in br1]

data_x = [
  0, 0, 0, bf_max_times[0], 0,
  0, 0, 0, 0, bf_max_times[1],
  0, 0, 0, 0, bf_max_times[2]
]
data_y = [
  0, 0, 0, bf_average_times[0], 0,
  0, 0, 0, 0, bf_average_times[1],
  0, 0, 0, 0, bf_average_times[2]
]

plt.title("(NR) BRUTE FORCE pro sady instancí velikosti: 4, 10, 15", fontweight = "bold", fontsize = 15)
plt.bar(br1, data_x, color = "r", width = barWidth, edgecolor = "grey", label = "BRUTE FORCE (maximální počet navštívených konfigurací)")
plt.bar(br2, data_y, color = "g", width = barWidth, edgecolor = "grey", label = "BRUTE FORCE (průměrný počet navštívených konfigurací)")
plt.xlabel("Velikost instance", fontweight = "bold", fontsize = 15)
plt.ylabel("Počet navštívených konfigurací", fontweight = "bold", fontsize = 15)
plt.xticks([r - barWidth/2 for r in range(0, 20, 5)], range(0, 20, 5))
 
plt.legend()
plt.show()
```
