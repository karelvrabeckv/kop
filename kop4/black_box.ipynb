```python
import matplotlib.pyplot as plt
import pandas as pd
import tester as t
```

```python
n40_result = t.run("n40.dat", 200, 10, 0.97, 40)
n37_result = t.run("n37.dat", 200, 10, 0.97, 37)
n35_result = t.run("n35.dat", 200, 10, 0.97, 35)
n32_result = t.run("n32.dat", 200, 10, 0.97, 32)
```

```python
inst_sizes = [32, 35, 37, 40]
avg_times = [n32_result[1], n35_result[1], n37_result[1], n40_result[1]]
max_times = [n32_result[2], n35_result[2], n37_result[2], n40_result[2]]
avg_errors = [n32_result[3], n35_result[3], n37_result[3], n40_result[3]]
max_errors = [n32_result[4], n35_result[4], n37_result[4], n40_result[4]]

plt.subplots(figsize = (12, 6))
plt.title("Zvislost kvality SA na parametru INST_SIZE", fontweight = "bold", fontsize = 15)
plt.xlabel("Velikost instance", fontweight = "bold", fontsize = 15)
plt.ylabel("Prmrn chyba", fontweight = "bold", fontsize = 15)
plt.plot(inst_sizes, avg_errors, marker = "o", linestyle = "solid", color = "black", label = "SA")
plt.grid()
plt.legend()
plt.show()

plt.subplots(figsize = (12, 6))
plt.title("Zvislost vpoetn nronosti SA na parametru INST_SIZE", fontweight = "bold", fontsize = 15)
plt.xlabel("Velikost instance", fontweight = "bold", fontsize = 15)
plt.ylabel("Prmrn as CPU (ms)", fontweight = "bold", fontsize = 15)
plt.plot(inst_sizes, avg_times, marker = "o", linestyle = "solid", color = "black", label = "SA")
plt.grid()
plt.legend()
plt.show()

df = pd.DataFrame({
  "Velikost instance": inst_sizes,
  "Prmrn chyba": [format(val, ".5f") for val in avg_errors],
  "Maximln chyba": [format(val, ".5f") for val in max_errors],
  "Prmrn as CPU (ms)": [format(val, ".5f") for val in avg_times],
  "Maximln as CPU (ms)": [format(val, ".5f") for val in max_times],
})
df.style.set_table_styles([{
  "selector": "th, td",
  "props": [("background-color", "white"), ("color", "black"), ("outline", "1px solid black"), ("font-size", "20px"), ("text-align", "center")]
}]).hide_index()
```
