```python
import matplotlib.pyplot as plt
import pandas as pd
import tester as t
```

```python
start_temps = [5, 10, 25, 50, 100, 200, 300, 400, 500, 1000]
start_temps_results = [t.run("n40.dat", start_temp, 1, 0.99, 40, 50) for start_temp in start_temps]
```

```python
avg_errors = [result[3] for result in start_temps_results]
avg_times = [result[1] for result in start_temps_results]

plt.subplots(figsize = (12, 6))
plt.title("Závislost kvality SA na parametru START_TEMP", fontweight = "bold", fontsize = 15)
plt.xlabel("Počáteční teplota", fontweight = "bold", fontsize = 15)
plt.ylabel("Průměrná chyba", fontweight = "bold", fontsize = 15)
plt.plot(start_temps, avg_errors, marker = "o", linestyle = "solid", color = "red", label = "SA")
plt.grid()
plt.legend()
plt.show()

plt.subplots(figsize = (12, 6))
plt.title("Závislost výpočetní náročnosti SA na parametru START_TEMP", fontweight = "bold", fontsize = 15)
plt.xlabel("Počáteční teplota", fontweight = "bold", fontsize = 15)
plt.ylabel("Průměrný čas CPU (ms)", fontweight = "bold", fontsize = 15)
plt.plot(start_temps, avg_times, marker = "o", linestyle = "solid", color = "red", label = "SA")
plt.grid()
plt.legend()
plt.show()

df = pd.DataFrame({
  "Počáteční teplota": start_temps,
  "Průměrná chyba": [format(val, ".5f") for val in avg_errors],
  "Průměrný čas CPU (ms)": [format(val, ".5f") for val in avg_times],
})
df.style.set_table_styles([{
  "selector": "th, td",
  "props": [("background-color", "white"), ("color", "black"), ("outline", "1px solid black"), ("font-size", "20px"), ("text-align", "center")]
}]).hide_index()
```
